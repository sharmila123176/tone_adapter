```python
import streamlit as st
import pandas as pd
from transformers import pipeline

# Load model pipeline
emotion_pipeline = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    return_all_scores=False
)

st.title("Emotion Detection App 🎭")

# Upload CSV
uploaded_file = st.file_uploader("Upload your CSV file", type=["csv"])

if uploaded_file:
    df = pd.read_csv(uploaded_file)
    st.write("Available columns:", df.columns.tolist())

    # Let user pick a column
    column_name = st.selectbox("Select the column containing text data", df.columns.tolist())

    # Analyze emotions
    if st.button("Detect Emotions"):
        results = []
        for text in df[column_name][:10]:  # Change 10 to len(df) for full
            prediction = emotion_pipeline(text)[0]
            results.append({
                "Response": text,
                "Detected_Tone": prediction['label'],
                "Confidence": round(prediction['score'], 2)
            })

        df_result = pd.DataFrame(results)
        st.write(df_result)
```
